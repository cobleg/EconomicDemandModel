# Introduction
This notebook estimates panel regression models for analysing the influence of economic variables on the demand for grid-delivered electricity. The models contained in this notebook represent long-run causal influences of the retail electricity price and economic output on the demand for grid-delivered electricity. The estimated models are of the following type:
$$Q_{it}^d= A_{it}P^{\beta_i}_{t}O^{\gamma_i}_{t}$$


where:
- $Q^d_{it}$ represents the demand for grid-delivered electricity for customer group $i$ in year $t$
- $A_{it}$ denotes all sources of influence on the variation in $Q_{it}^d$ except for $P_t$ and $O_t$; 
- $i$ represents a customer group from the set $i=\{Residential, Small Commercial, Large Commercial, Transmission, Other\}$; 
- $t$ denotes a time index (in years), $t=\{2008, 2009, ...,2019\}$; 
- $P$ is the electricity price;
- $O$ is the level of economic output; and
- the exponents $\beta$ and $\gamma$ represent iso-elastic parameters to be estimated.



In general, elasticity is defined as:
$$e=\frac{\% \Delta Y}{ \% \Delta X}$$
where:
- $X$ is the set of input variables causing a change, in this case $X={P, O}$
- $Y$ is the set of output variables, in this case $Y=\{Q_i\}$

In words, elasticity measures the percent change in a variable $Y$ given a percent change in a variable $X$. There can be either an inelastic relationship between $X$ and $Y$ (where $|e|<1$) or an elastic relationship (where $|e|>1$).



##### More on $A_{it}$
The variable $A_{it}$ is a mix of unobserved variables. Unobserved variables cannot be directly controlled for and can lead to bias in the estimated parameters. So indirect controls need to be used to minimise the bias effect, such as:
- Binary variables (aka fixed effects) that control for systematic changes across groups of subjects (in this case, customer groups and/or spatial regions).
- Random effects. See this wikipedia article for more information; <a href="https://en.wikipedia.org/wiki/Random_effects_model">Random Effects</a>.



##### Impact of changing technology on electricity demand
The variation attributed to unobserved variables can be segmented into a variety of influences including changing technology. Note that in economic theory, technology refers to the rules (or know-how) that describe how to transform the inputs of production into outputs. In application to electricity demand, technology determines the amount of electricity required in combination with other production inputs such as physical capital (e.g. electrical appliances) and labour services (e.g. someone using a computer) to produce a given output (e.g. an econometric analysis of electricity demand). Changing technology can either increase or decrease the demand for electricity. 



##### Two pathways of influence of technology

The two pathways are:
1. Directly via an alternative source of energy supply (e.g. on-site electricity generation)
2. Indirectly via electricity-using electrical appliances and energy consumption offsets such as improved thermal insulation in buildings.




In recent years, new buildings and electrical appliances have been designed to use less energy (including electricity) than older buildings and electrical appliances. The aggregate impact of technological change depends on the rate of renewal in the stock of buildings and electrical appliances within the electricity grid supply zone. NOTE: The rate of renewal is likely to be correlated to economic cycles since increasing real wages and profits during an economic upswing tends to stimulate increased purchases of new buildings and electrical appliances.


One way of isolating the electricity-conserving effect of technological change via buildings is to compare the electricity demand of new buildings to older buildings that are similar (ideally identical) in all respects except for thermal efficiency. 
This can be achieved via hedonic indexing of building stocks.

For more information about energy efficiency trends in Australia's building stock, please see this document:

<a href="https://www.aemo.com.au/-/media/Files/Electricity/NEM/Planning_and_Forecasting/NEM_ESOO/2018/Energy-Efficiency-Impacts-on-Electricity-and-Gas-Demand-to-2037-38-by-Strategy-Policy-Research.pdf">AEMO(2019). Energy Efficiency Impacts on Electricity and Gas Demand to 2037-38.</a>

The AEMO energy efficiency paper suggests aggregate rate of energy efficiency improvement is:
- Approximately 2.5% per year in the commercial building sector for the period 2001-2015; and
- Approximately 3.2% per year in the residential building sector for the same period.

NOTE: In applying these numbers to aggregate (e.g. system-level) electricity demand, we need to apply weights. For example, electricity use for space heating and cooling would account for approximately 60% of total electricity demand in homes, which accounts for approximately 45% of system-level demand. Hence, the aggregate effect of improving thermal efficiency in residential buildings would be: 3.5% x .6 x .45 = 1.0% per year (approximately).



### Economic theory of demand
This analysis draws on economic theory to develop formal expectations about the expected signs of economic variables (e.g. prices, income, production output). Specifically, we expect that there is:
- an inverse (negative) relationship between the price of grid-delivered electricity and the quantity of electricity demanded; and
- a direct (positive) relationship between the level of economic output and the demand for electricity. 

The magnitude of the change in level of price and economic output on grid-delivered electricity demand are derived from data rather than theory and so need to be estimated.



#### Factoring in solar PV generated electricity
It can be shown that adoption of solar PV systems in recent years is largely the result of the increasing price of grid-delivered electricity. That is, assuming that an electricity consumer wishes to maintain a constant level of electricity expenditure, the consumer can achieve this objective by increasing the supply of solar PV generated electricity and reducing the supply of grid-delivered electricity. 

That is:
$$E_{Total}=E_{G}+E_{PV}$$

where $E_{Total}$ represents expenditure on electricity, $E_{G}$ represents expenditure on grid-delivered electricity and $E_{PV}$ represents solar PV generated electricity. Now consider changing the amount spent on each supply of electricity so that there is no change to total expenditure, i.e. $\Delta E_{Total}=0$. This implies

$$\Delta E_{Total}=\Delta E_{G}+\Delta E_{PV}=0$$

Restating this in terms of price and quantity we have

$$ \Delta P_{G} Q_{G} + P_{G} \Delta Q_{G} + \Delta P_{PV} Q_{PV} + P_{PV}\Delta Q_{PV} = 0 $$

Imposing a change in the price of network-delivered electricity, i.e. $\Delta P_{G} <>0 $ implies offsetting changes in the other variables is required to satisfy the constraint of no change to total electricity consumption. If $\Delta P_{PV}$ does not change, then the offsetting change must occur via the quantities $Q_{G}$ and $Q_{PV}$ implying that $\Delta Q_{G} <>0 $ and $\Delta Q_{PV} <>0$, which implies

$$ \Delta P_{G} Q_{G} + P_{G} \Delta Q_{G} + P_{PV}\Delta Q_{PV} = 0 $$

Let $ P_{G}=1$, $ P_{PV}=1$ and $Q_{G}=1$, this simplifies to

$$ \Delta P_{G} + \Delta Q_{G} + \Delta Q_{PV} = 0 $$


Rearranging this, we can see that the change in demand for electricity is opposite in sign to the change in the price of grid-delivered electricity

$$ \Delta Q_{PV} + \Delta Q_{G} = -\Delta P_{G} $$

So now we can see that if $\Delta P_{G}>0$, the total change in demand for electricity must decrease. Now isolating the change in grid-delivered electricity, we see that 

$$ \Delta Q_{G} = -\Delta P_{G} - \Delta Q_{PV} $$

Recognising that if solar generated electricity is always at maximum production, then $\Delta Q_{PV}=0$ and , 

$$ \Delta Q_{G} = -\Delta P_{G} $$


This analysis suggests that grid-delivered electricity will decrease when the electricity price increases. Note that, in this case, the size of the reduction is increased by solar PV electricity generation technology. Moreover, solar PV adoption is part of the price response (realised via the substitution effect to keep expenditure within budget). So no need to add a separate control for solar PV unless you're interested in identifying the price response of the excess demand beyond what solar PV generation technology can provide.

## Panel regression
Panel regression is a statistical method that controls for unobserved variation across the subjects being studied. In this case, we are exploring the effects of electricity prices and income on variation in the demand for grid-delivered electricity. Panel regression helps reduce bias induced by omitting unobserved variables as well as aggregation bias.

There are various types of panel regression:
1. Fixed effects
2. Random effects

See this article for a basic introduction: https://towardsdatascience.com/understanding-panel-data-regression-c24cd6c5151e

```r
library(plm) # load the panel regression procedure
```

### Model for customer class: Residential
Results below indicate that residential electricity demand is influenced by electricity price and average economic output.Coefficients suggest that:
1. An autonomous demand for grid-delivered electricity of 14 kWh per day per residence (i.e. house/home)
2. 10% increase in the retail electricity price results in a 1.5% decrease in the demand for grid-delivered electricity
3. A 10% increase in average economic output results in a 9.4% increase in the demand for grid-delivered electricity.

```r
residential.random <- plm(EXPORT_PER_NMI ~ Electricity_Real+SWIS_GRP_POP, data=data2[ which( data2$CUST_GROUP=="RESIDENTIAL"),], index=c("LOAD_AREA_NAM", "YEAR"), model="random")
summary(residential.random)
```

### Model for customer class: Small Commercial
Results below indicate that average economic output is a key driver of demand for electricity from this customer class. The results indicate:
1. No autonomous demand for grid-delivered electricity 
2. A 10% increase in the retail electricity price causes a 0% to 1.2% decrease in grid-delivered electricity demand. Note the t-statistic is not significantly different to zero at conventional levels of significance.
3. A 10% increase in average economic output causes a 29% increase in grid-delivered electricity demand.

```r
small_commercial.random <- plm(EXPORT_PER_NMI ~ Electricity_Real+SWIS_GRP_POP, data=data2[ which( data2$CUST_GROUP=="SMALL COMMERCIAL"),], index=c("LOAD_AREA_NAM", "YEAR"), model="within")
summary(small_commercial.random)
```

```r
small_commercial.random <- plm(EXPORT_PER_NMI ~ Electricity_Real+SWIS_GRP_POP, data=data2[ which( data2$CUST_GROUP=="SMALL COMMERCIAL"),], index=c("LOAD_AREA_NAM", "YEAR"), model="random")
summary(small_commercial.random)
```

```r
small_commercial.random <- plm(EXPORT_PER_NMI ~ Electricity_Real+SWIS_GRP_POP, data=data2[ which( data2$CUST_GROUP=="SMALL COMMERCIAL"),], index=c("LOAD_AREA_NAM", "YEAR"), model="pooling")
summary(small_commercial.random)
```
